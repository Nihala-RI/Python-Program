#Creating Bars

With Pyplot, you can use the bar() function to draw bar graphs:

A bar plot or bar chart is a graph that represents the category of data with rectangular bars with lengths and heights that is proportional to the values which they represent. The bar plots can be plotted horizontally or vertically. A bar chart describes the comparisons between the discrete categories. One of the axis of the plot represents the specific categories being compared, while the other axis represents the measured values corresponding to those categories.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

#Pie chart

A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data. Pie charts are commonly used in business presentations like sales, operations, survey results, resources, etc. as they provide a quick summary.

Syntax: matplotlib.pyplot.pie(data, explode=None, labels=None, colors=None, autopct=None, shadow=False)

Parameters:

data represents the array of data values to be plotted, the fractional area of each slice is represented by data/sum(data) .

 If sum(data)<1, then the data values returns the fractional area directly, thus resulting pie will have empty wedge of size 1-sum(data).

labels is a list of sequence of strings which sets the label of each wedge.
color attribute is used to provide color to the wedges.

autopct is a string used to label the wedge with their numerical value.
shadow is used to create shadow of wedge.

```python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np


# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

# Creating plot
fig = plt.figure(figsize=(10, 7))
plt.pie(data, labels=cars)

# show plot
plt.show()
```

By default the plotting of the first wedge starts from the x-axis and moves counterclockwise:



The startangle parameter is defined with an angle in degrees, default angle is 0:

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data to plot
data = [0.2, 0.3, 0.5]

# Labels for each slice
labels = ['Apples', 'Bananas', 'Cherries']

# Colors for each slice
colors = ['red', 'yellow', 'purple']

# Create the pie chart
plt.pie(
    data,                     # The data values to be plotted
    explode=[0.1, 0, 0],      # Explode the first slice (Apples) slightly
    labels=labels,            # Set the labels for the wedges
    colors=colors,            # Set the custom colors for each slice
    autopct='%1.1f%%',        # Display the percentage on each slice
    shadow=True               # Add a shadow to the pie chart
)

# Display the pie chart
plt.title('Fruit Distribution')
plt.show()
```

Using Stacking in Two Directions

```python
#1-D Array of Subplots
import matplotlib.pyplot as plt

# Create a 1x3 grid of subplots (1 row, 3 columns)
fig, ax = plt.subplots(1, 3, figsize=(10, 3))

# Plot on each subplot
ax[0].plot([1, 2, 3], [1, 4, 9], 'r')  # Red line
ax[0].set_title("Plot 1")

ax[1].plot([1, 2, 3], [1, 2, 3], color='g')  # Green line
ax[1].set_title("Plot 2")

ax[2].plot([1, 2, 3], [9, 4, 1], 'b')  # Blue line
ax[2].set_title("Plot 3")

# Adjust layout to avoid overlap
plt.tight_layout()
plt.show()
```

```python
##2-D Array of Subplots
import matplotlib.pyplot as plt

# Create a 2x2 grid of subplots
fig, ax = plt.subplots(2, 2, figsize=(8, 6))


# Plot on each subplot
ax[0, 0].plot([1, 2, 3], [1, 4, 9], 'r')  # Top-left
ax[0, 0].set_title("Top Left")

ax[0, 1].plot([1, 2, 3], [1, 2, 3], 'g')  # Top-right
ax[0, 1].set_title("Top Right")

ax[1, 0].plot([1, 2, 3], [9, 4, 1], 'b')  # Bottom-left
ax[1, 0].set_title("Bottom Left")

ax[1, 1].plot([1, 2, 3], [3, 3, 3], 'k')  # Bottom-right
ax[1, 1].set_title("Bottom Right")

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#Stacking in 2-D

# Import necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Simulated real-life weather data (for 7 days)
days = np.array([1, 2, 3, 4, 5, 6, 7])  # Days of the week
temperature = np.array([22, 25, 23, 26, 27, 24, 22])  # Daily temperature (°C)
humidity = np.array([55, 60, 58, 65, 62, 59, 57])  # Daily humidity (%)
wind_speed = np.array([5, 8, 7, 6, 9, 4, 3])  # Wind speed (km/h)
precipitation = np.array([0, 2, 0, 5, 3, 1, 0])  # Precipitation (mm)

# Create 6 subplots in a 3x2 grid
fig, ((ax1, ax2),(ax3, ax4)) = plt.subplots(2, 2, figsize=(10, 10))

# Plot temperature data
ax1.plot(days, temperature, color="orange", marker='o')
ax1.set_title("Temperature Over a Week (°C)")
ax1.set_xlabel("Days")
ax1.set_ylabel("Temperature (°C)")

# Plot humidity data
ax2.plot(days, humidity, color="green", marker='o')
ax2.set_title("Humidity Over a Week (%)")
ax2.set_xlabel("Days")
ax2.set_ylabel("Humidity (%)")

# Plot wind speed data
ax3.plot(days, wind_speed, color="blue", marker='o')
ax3.set_title("Wind Speed Over a Week (km/h)")
ax3.set_xlabel("Days")
ax3.set_ylabel("Wind Speed (km/h)")

# Plot precipitation data
ax4.bar(days, precipitation, color="magenta")
ax4.set_title("Precipitation Over a Week (mm)")
ax4.set_xlabel("Days")
ax4.set_ylabel("Precipitation (mm)")

# Set a single title for the entire figure
fig.suptitle('Temperature', fontsize=16)

# Adjust layout to prevent overlap between subplots
plt.tight_layout()

# Show the plot
plt.show()
```

##Stacked bar chart with more than 2 data

```python
# importing package
import matplotlib.pyplot as plt
import numpy as np

# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])

# plot bars in stack manner
plt.bar(x, y1, color='r')
plt.bar(x, y2, bottom=y1, color='b',)
plt.bar(x, y3, bottom=y1+y2, color='y')
plt.bar(x, y4, bottom=y1+y2+y3, color='g')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.legend(["Round 1", "Round 2", "Round 3", "Round 4"])
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = {'START': ['Fort Pierce', 'Fort Pierce', 'Cary', 'Cary', 'Jamaica'],
        'STOP': ['Fort Pierce', 'Fort Pierce', 'Morrisville', 'Cary', 'New York'],
        'MILES': [5.1, 4.8, 8.3, 0.8, 16.5],
        'PURPOSE': ['Meal/Entertain', 'Errand/Supplies', 'Meeting', 'Meeting', 'Customer Visit']}

df = pd.DataFrame(data)

# Pivot the data for a stacked bar chart (rows as START-STOP, columns as PURPOSE)
pivot_df = df.pivot_table(values='MILES', index=['START', 'STOP'], columns='PURPOSE', fill_value=0)
pivot_df
```

```python
# Plot the stacked bar chart
pivot_df.plot(kind='bar', stacked=True, figsize=(10, 6))
plt.title('Miles Traveled by Purpose for Each Start-Stop Location')
plt.xlabel('Start-Stop Locations')
plt.ylabel('Miles Traveled')
plt.xticks(rotation=45, ha="right")  # Rotate x labels for readability
plt.tight_layout()
plt.show()
```
